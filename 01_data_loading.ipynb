```python
import pandas as pd

metadata = pd.read_csv("../data/raw/metadata.csv")
metadata.head()
```

```python
metadata.shape
```

```python
metadata['site_id'].unique()
```

```python
metadata['site_id'].value_counts()
```

```python
panther_buildings = metadata[metadata['site_id'] == 'Panther']
panther_buildings['primaryspaceusage'].value_counts()
```

```python
electricity = pd.read_csv("../data/raw/electricity_cleaned.csv")
electricity.shape
```

```python
electricity.columns[:10]
```

```python
panther_building_ids = panther_buildings['building_id'].tolist()
len(panther_building_ids)
```

```python
columns_to_keep = ['timestamp'] + panther_building_ids
electricity_panther = electricity[columns_to_keep]
electricity_panther.shape
```

```python
available_columns = electricity.columns.tolist()
panther_ids_available = [b for b in panther_building_ids if b in available_columns]
len(panther_ids_available)
```

```python
columns_to_keep = ['timestamp'] + panther_ids_available
electricity_panther = electricity[columns_to_keep]
electricity_panther.shape
```

```python
electricity_panther.head()
```

```python
electricity_long = electricity_panther.melt(
    id_vars='timestamp',
    var_name='building_id',
    value_name='meter_reading'
)
electricity_long.shape
```

```python
electricity_long.head(10)
electricity_long['building_id'].nunique()
```

```python
electricity_long['meter_reading'].isna().mean()
```

```python
missing_per_building = electricity_long.groupby('building_id')['meter_reading'].apply(lambda x: x.isna().mean())
missing_per_building.sort_values(ascending=False).head(15)
```

```python
weather = pd.read_csv("../data/raw/weather.csv")
weather.shape
weather.head()
```

```python
weather_panther = weather[weather['site_id'] == 'Panther']
weather_panther.shape
```

```python
merged = electricity_long.merge(weather_panther, on='timestamp', how='left')
merged.shape
merged.head()
```

```python
merged.shape
merged.columns.tolist()
```

```python
final = merged.merge(panther_buildings, on='building_id', how='left')
final.shape
final.columns.tolist()
```

```python
final = final.drop(columns=['site_id_y'])
final = final.rename(columns={'site_id_x': 'site_id'})
final.shape
```

```python
final.to_csv("../data/processed/panther_merged.csv", index=False)
```
